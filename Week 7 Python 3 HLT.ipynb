```python
fruit=str(input("give me a name of  a fruit ")).lower() #convert all letters to lowercase
count_vowel=0 
b=['a','e','i','o','u']
for i in fruit: #for any letter in fruit
  if i in b: # check the letter is a vowel
    count_vowel+=1 #if yes,add 1 to vowel counter
  else: #else carry on
    pass
print(count_vowel, " vowels are in "+ str(fruit)) #prints how many vowels is in the given fruit
```
